```python
class LogisticRegression(LinearModel):
    """Logistic regression with Newton's Method as the solver.

    Example usage:
        > clf = LogisticRegression()
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def fit(self, x, y):
        """Run Newton's Method to minimize J(theta) for logistic regression.

        Args:
            x: Training example inputs. Shape (m, n).
            y: Training example labels. Shape (m,).
        """
        # *** START CODE HERE ***
        def h(theta, x):
            z = np.dot(x, theta) 
            return 1 / (1 + np.exp(-z)) # shape (m,)
        
        def gradient(theta, x, y):
            m = x.shape[0]
            return -1 / m * np.dot(x.T, y - h(theta, x)) # shape (n,)
        
        def Hessian(theta, x, y):
            m = x.shape[0]
            h_theta_x = h(theta, x)
            diag = h_theta_x * (1 - h_theta_x)
            return 1 / m * np.dot(np.dot(x.T, np.diag(diag)), x) # shape (n, n)
        
        def update_theta(theta, x, y):
            grad = gradient(theta, x, y) # shape (n,)
            H = Hessian(theta, x, y)
            return theta - np.dot(np.linalg.inv(H), grad)
        
        # Initialize theta, if not in the attribute
        m, n = x.shape
        if self.theta is None:
            self.theta = np.zeros(n)
        # Keep updating theta with Newton's Method until reaching precision
        while True:
            new_theta = update_theta(self.theta, x, y)
            if np.linalg.norm(new_theta - self.theta, 1) < self.eps: 
                break
            self.theta = new_theta
        return self.theta
        # *** END CODE HERE ***

    def predict(self, x):
        """Make a prediction given new inputs x.

        Args:
            x: Inputs of shape (m, n).

        Returns:
            Outputs of shape (m,).
        """
        # *** START CODE HERE ***
        prob = 1 / (1 + np.exp(-x @ self.theta)) # shape (m,)
        return prob >= 0.5 # return true if greater than 0.5
        # *** END CODE HERE ***
```

Then we train the model and plot the decision boundary of the training set:

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
util.plot(x_train, y_train, theta=log_reg.theta)

print(f"Theta is: {log_reg.theta}")
print(f"The accuracy on training set is: {np.mean(log_reg.predict(x_train) == y_train)}")
```

Also plot the model for the validation set:

```python
util.plot(x_valid, y_valid, log_reg.theta)
print(f"The accuracy on validation set is: {np.mean(log_reg.predict(x_valid) == y_valid)}")
```
